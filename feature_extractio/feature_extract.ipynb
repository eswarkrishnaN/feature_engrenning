ft.EntitySet: This is the class from the Featuretools library that is used to create an EntitySet object.

id="data": The id parameter is used to provide a unique identifier for the EntitySet. In this case, the identifier is set as "data"

```python
es = ft.EntitySet(id="data")
```

dataframe=df: This parameter specifies the DataFrame (df) that you want to add to the EntitySet. It represents the data that will be used to create the new entity.

dataframe_name="data": This parameter sets the name for the new entity within the EntitySet. In this case, the name is set as "data".

index="rows": This parameter indicates the type of index to use for the new entity. Setting it to "rows" means that each row in the DataFrame will be treated as a separate observation or instance.

make_index=True: This parameter specifies whether to create a new index column for the entity if one doesn't exist in the DataFrame. Setting it to True means that Featuretools will generate a new index column if needed.

time_index="invoice_date": This parameter specifies the column in the DataFrame that represents the time index or timestamp information. In this case, the "invoice_date" column is designated as the time index for the new entity.

logical_types={"customer_id": Categorical}: This parameter allows you to define the logical types of specific columns in the DataFrame. In this example, the "customer_id" column is assigned the logical type "Categorical", indicating that it represents a categorical variable

```python
es = es.add_dataframe(

        dataframe=df,

        dataframe_name="data",

        index="rows",

        make_index=True,

        time_index="invoice_date",

        logical_types={ "customer_id": Categorical},

)

es
```

base_dataframe_name="data": This parameter specifies the name of the base dataframe or entity on which the normalization will be applied. In this case, the base dataframe is named "data".

new_dataframe_name="invoices": This parameter sets the name for the new normalized dataframe or entity that will be created. The normalized dataframe represents a specific relationship or aspect extracted from the base dataframe. Here, it is named "invoices".

index="invoice": This parameter indicates the column in the base dataframe that serves as the unique identifier or index for the new normalized dataframe. It represents the relationship between the base dataframe and the new normalized dataframe. In this case, the "invoice" column is designated as the index.

copy_columns=["customer_id"]: This parameter specifies the columns that will be copied from the base dataframe to the new normalized dataframe. These columns represent the shared attributes or properties between the two dataframes. Here, the "customer_id" column is selected to be copied.

```python
es.normalize_dataframe(

        base_dataframe_name="data",

        new_dataframe_name="invoices",

        index="invoice",

        copy_columns=["customer_id"],

)
```

Creating features with general and cumulative operations

Featuretools uses what are called transform primitives to create features. Transform primitives take one or more columns in a dataset as input and return one or more columns as output. They are applied to a single dataframe.

Featuretools divides its transform primitives into various categories depending on the type of operation they perform or the type of variable they modify. For example, general transform primitives apply mathematical operations, such as the square root, the sine, and the cosine. Cumulative transform primitives create new features by comparing a row’s value to the previous row’s value. For example, the cumulative sum, cumulative mean, and cumulative minimum and maximum values belong to this category, as well as the difference between row values. There is another cumulative transformation that can be applied to datetime variables, which is the time since previous, which determines the time passed between two consecutive timestamps

```python
import numpy as np
import pandas as pd

df = pd.read_csv("retail.csv",parse_dates=["invoice_date"])

#caputering the two variables 
numeric_vars = ["quantity","price"]

#caputer the names of the cumulative fucntions in a list
func = ["cumsum","cummax","diff"]

#creating a new names for the variables 
new_names = [f"{var}_{function}" for function in func
             for var in numeric_vars]

new_names
```

The previous step returns the cumulative sum or cumulative maximum value, or the difference between rows, within each invoice. As soon as it encounters a new invoice number, it starts afresh. To better understand the result, execute df[df["invoice"] == "489434" ][numeric_vars + new_names] and compare the result to that of df[df["invoice"] == "489435" ][numeric_vars + new_names]

```python
#create a list with names for the new variables
new_names = [f"{var}_{function}" for function in ["sin", "cos"] for var in numeric_vars]


df[new_names] = df[numeric_vars].agg([np.sin, np.cos])

df[new_names]
```

Combining numerical features

 we saw that we can create new features by combining variables with mathematical operations. Featuretools supports a number of operations to combine variables, including addition, division, modulo, and multiplication. In this recipe, we will learn how to combine these features with Featuretools

```python
import pandas as pd

import featuretools as ft

from woodwork.logical_types import Categorical

df = pd.read_csv("retail.csv", parse_dates=["invoice_date"])

#set up an entity set 
es = ft.EntitySet(id="data")


#add the dataframe to the entity set
es = es.add_dataframe(

        dataframe=df,

        dataframe_name="data",

        index="rows",

        make_index=True,

        time_index="invoice_date",

        logical_types={"customer_id": Categorical},

)

#create a new dataframe with a relationship to the dataframe

es.normalize_dataframe(

        base_dataframe_name="data",

        new_dataframe_name="invoices",

        index="invoice",

        copy_columns=["customer_id"],

)
es
```

```python
#multiply the quantity and price variables, which reflect the number of items bought and the unit price, respectively, to obtain the total amount pai

feature_matrix, feature_defs = ft.dfs(

        entityset=es,

        target_dataframe_name="data",

        agg_primitives=[],

        trans_primitives=["multiply_numeric"],

        primitive_options={

            ("multiply_numeric"): {

                'include_columns': {

                    'data': ["quantity", "price"]

                }

            }

        },

        ignore_dataframes=["invoices"],

)
feature_matrix
```

Extracting features from date and time

Extracting Features from Date and Time, we created various features from the date and time parts of datetime variables using pandas and Feature-engine. We can also extract multiple features from datetime variables, automatically utilizing Featuretools.

Featuretools supports the creation of various features from datetime variables using its datetime transform primitives. These primitives include the creation of features such as year, month, and day, and also other features such as is it lunch time or is it weekday?

```python
import pandas as pd
import featuretools as ft
from woodwork.logical_types import Categorical
```

```python
df = pd.read_csv("retail.csv", parse_dates=["invoice_date"])


#set up an entity set
es = ft.EntitySet(id="data")

#add the dataframe to the entity set
es = es.add_dataframe(

        dataframe=df,

        dataframe_name="data",

        index="rows",

        make_index=True,

        time_index="invoice_date",

        logical_types={"customer_id": Categorical},

)

# create a new dataframe with a relationship to the dataframe
es.normalize_dataframe(

        base_dataframe_name="data",

        new_dataframe_name="invoices",

        index="invoice",

        copy_columns=["customer_id"],

)

#make a list with string names that identify the features we want to create
date_primitives = ["day", "year", "month", "weekday",

                       "days_in_month", "part_of_day",

                       "is_federal_holiday",

                       "hour", "minute"]

feature_matrix, feature_defs = ft.dfs(

        entityset=es,
    
        target_dataframe_name="invoices",

        agg_primitives=[],

        trans_primitives=date_primitives,

)

feature_matrix
```

```python
from featuretools.primitives import DistanceToHoliday

#repare the primitive by specifying that we are interested in the UK’s bank holidays and that in particular, we want to know how long it is until Boxing Day


distance_to_boxing_day = DistanceToHoliday(holiday="Boxing Day", country="UK")

feature_matrix, feature_defs = ft.dfs(

        entityset=es,

        target_dataframe_name="invoices",

        agg_primitives=[],

        trans_primitives=[distance_to_boxing_day],

        verbose=True,

)

feature_defs

feature_matrix.head(20)
#that we do not have the price or description variables, among other things, because we transformed the invoices dataset in step 2 and not the entire dataframe
```
