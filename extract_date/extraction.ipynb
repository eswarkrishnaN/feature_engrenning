```python
import  pandas as pd
import numpy as np



rng_ = pd.date_range("2019-03-05",periods=20 , freq="D")

data = pd.DataFrame({"date":rng_})

```

```python
#we can extract the year
data["year"] = data["date"].dt.year

data["quarter"] = data["date"].dt.quarter


data["semester"] = np.where(data["quarter"]<3, 1, 2)

data
```

```python
data["month"] = data["date"].dt.month

data[["date", "month"]].head()
```

```python
data["week"] = data["date"].dt.isocalendar().week
data
```

```python
data["day_mo"] = data["date"].dt.day

data[["date", "day_mo"]].head()
```

```python
data["day_week"] = data["date"].dt.dayofweek

data[["date", "day_mo", "day_week"]].head()
```

```python
data["is_weekend"] = (data["date"].dt.dayofweek > 2).astype(int)

data[["date", "day_week", "is_weekend"]].head()
```

Extracting features from time with pandas

pandas.Series.dt.hour

pandas.Series.dt.minute

pandas.Series.dt.second

```python
import numpy as np
import pandas as pd

#it will set automatacally adding one year
rng_ = pd.date_range("2019-03-05",periods=20, freq ="1h15min10s")

df = pd.DataFrame({"date":rng_})

df.head()
```

```python
df["hour"] = df["date"].dt.hour

df["min"] = df["date"].dt.minute

df["sec"] = df["date"].dt.second

df.head()
```

```python
df[["h", "m", "s"]] = pd.DataFrame([(x.hour, x.minute,x.second) for x in df["date"]])

df.head()
```

```python
df["is_morning"] = np.where((df["hour"] < 12) & (df["hour"] > 6), 1, 0 )

df.head()


#pandas.Series.dt.microsecond

#pandas.Series.dt.nanosecond
```

```python
#work with different time zones, first, we must unify the time zone to the central time zone setting, utc = True
df['time1_utc'] = pd.to_datetime(df['time1'], utc=True)

df['time2_utc'] = pd.to_datetime(df['time2'], utc=True)

df
```

```python
df['time2'] = pd.concat([pd.Series(pd.date_range(start='2015-07-01 09:00', freq='H',
                                                 periods=3,tz='Europe/Berlin')),
                         pd.Series(pd.date_range(start='2015-08-01 09:00', freq='H',
                                                 periods=3,tz='US/Central'))], axis=0)

df
```

```python
df['elapsed_days'] = (df['time2_utc'] - df['time1_utc']). dt.days

df['elapsed_days'].head()
```

Automating feature extraction with Feature-engine

```python
#it is just easy you can extract the data using feature-engine

from feature_engine.datetime import DatetimeFeatures

rng_ = pd.date_range('2019-03-05', periods=20, freq='D')

data = pd.DataFrame({'date': rng_})

dtfs = DatetimeFeatures(variables=None,features_to_extract=None,)

dft = dtfs.fit_transform(data)

vars_ = [v for v in dft.columns if "date" in v]

dft[vars_].head()
```
